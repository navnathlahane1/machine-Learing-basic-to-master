```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
data["sex"] = enc.fit_transform(data['sex'].astype('str'))
data["smoker"] = enc.fit_transform(data['smoker'].astype('str'))
data["day"] = enc.fit_transform(data['day'].astype('str'))
data["time"] = enc.fit_transform(data['time'].astype('str'))



```

```python
from sklearn.cluster import KMeans
```

```python
ideal_k = []
for i in range(1,21):
    est_kmeans = KMeans(n_clusters=i,random_state=0)
    est_kmeans.fit(data)
    ideal_k.append([i,est_kmeans.inertia_])
```

```python
ideal_k = np.array(ideal_k)
```

```python
plt.plot(ideal_k[:,0],ideal_k[:,1])
plt.show
```

```python
est_kmeans=KMeans(n_clusters=5,random_state=2)
est_kmeans.fit(data)
pred_kmeans = est_kmeans.predict(data)
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=pred_kmeans)
plt.show()

```

```python
from sklearn.cluster import MeanShift


```

```python
est_Meanshift = MeanShift()
est_Meanshift.fit(data)
pred_meanshift = est_Meanshift.predict(data)
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=pred_meanshift)
plt.show()

```

** DBSCAN Algorithm
* A core point: A point that has at least the minimum number of dataa points within its eps radius.
* A border point: A point that is within the esp radius of a core point, but does not have the required number of data points within its own radius .
* A noise point : All points that do not meet the preceding descriptions.


```python
from sklearn.cluster import DBSCAN
```

```python
est_dbscan = DBSCAN(eps=0.1)
pred_dbscan= est_dbscan.fit_predict(data)
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=pred_dbscan)
plt.show()
```

```python
from sklearn.metrics import homogeneity_score
score = homogeneity_score(labels_pred=data,labels_true=data)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
```

```python
KMeans_score = silhouette_score(data,pred_kmeans )
MeanShift_score = silhouette_score(data, pred_meanshift)
#dbscan_score = silhouette_score(data,pred_dbscan)
print(KMeans_score,MeanShift_score)

```

```python
kmeansscore = calinski_harabasz_score(data,pred_kmeans)
MeanShiftscore = calinski_harabasz_score(data,pred_meanshift)
#dbscanscore = calinski_harabasz_score(data,pred_dbscan)
print(kmeansscore,MeanShiftscore)
```
