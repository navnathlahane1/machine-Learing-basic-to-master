```python
import pandas as pd 
import pickle 
import os
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv("output.csv",header=None)
```

```python
data.head(50
          
          )
```

```python
x = data.iloc[:,:9]
y = data.iloc[:,9]
```

```python
model = MLPClassifier(max_iter=1200)
model.fit(x,y)
```

```python
path = os.getcwd() + "/model_exercise.pkl"
file = open(path,"wb")
pickle.dump(model,file)
```
